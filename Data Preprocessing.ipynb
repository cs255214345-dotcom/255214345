```python
import pandas as pd
from pandas import DataFrame
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.feature_selection import VarianceThreshold
data={'name':['Rex','Peter','Sam','Rita' ,'Hilda'],
'age':[1,95,22,0,60],
'height':[20,180,160,145,125],
'weight':[0,100,75,55,40],
'bp':[80,100,65,70,40]}
frame=DataFrame(data)
print(frame)
frame.to_csv('patient_data.csv',index=False)
df=pd.read_csv('patient_data.csv')
print(df)
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].replace(0, df[numeric_cols].mean())
print(df)
frame2=df[['age','bp','height','weight']]
print(frame2)
scaler = StandardScaler()
scaler.fit(frame2)
scaled=scaler.transform(frame2)
print(scaled)
scaled.shape
print(&quot;mean:%s&quot;%scaled.mean(axis=0))
print(&quot;Standard deviation:%s&quot;%scaled.std(axis=0))
frame3=df[['age','bp','height','weight']]
min_max = MinMaxScaler()
min_max.fit(frame3)
scaled2=min_max.transform(frame3)
print(scaled2)
print(&quot;mean:%s&quot;%scaled2.mean(axis=0))
print(&quot;Standard deviation:%s&quot;%scaled.std(axis=0))
sel=VarianceThreshold(threshold=(.5*(1-.9)))
useful_features1=sel.fit_transform(scaled)
print('Useful feature values based on standard scaler\n')
print(useful_features1)
useful_features2=sel.fit_transform(scaled2)
print('\nUseful features values based on MinMaxScaler')
print(useful_features2)
```
